```python
def missing_attr_stats(da):
    l=[]
    for f in da.features_with_nan:
        if f in missing_feature_map.keys():
            count,percent = da.find_missing_count(f)
            print("{},  {} , {}  ".format(f,count,percent))
        else:
            l.append((f,da.find_missing_count(f)))
    return l

def features_with_all_missing(da):
    ignore_list = missing_attr_stats(da)
    to_drop = []
    for stat in ignore_list:
        attr, s = stat
        percent = s[1]
        if percent == 100.0:
            to_drop.append(attr)
    return to_drop
```

```python
def fill_missing_attr(da):
    print("======================================")
    print("Before filling the values")
    print("======================================")
    missing_attr_stats(da)
    for f in da.features_with_nan:
        if f in missing_feature_map.keys():
            da.fill_missing(f)
    print("======================================")
    print("After filling the values")
    print("======================================")
    cant_fill = missing_attr_stats(da)
    print()
    print("======================================")
    print("Features which could not be filled")
    print("Feature Name    |  Missing value count   | Percentage   ")
    for stat in cant_fill:
        feature, s = stat
        print("{}\t\t{}\t\t{}".format(feature, s[0],s[1]))
```

```python
fill_missing_attr(state)
```

```python
fill_missing_attr(county)
```
